```python
import numpy as np
from os import listdir
import fnmatch

from os.path import isfile, join, isdir
import pandas as pd
def get_alpha(filename):
    for chunk in filename.replace(".npy", "").split("_"):
        if chunk.startswith('priv'):
            alpha = float(chunk.replace("priv", ""))/10
    return(alpha)

def get_k(filename):
    for chunk in filename.replace(".npy", "").split("_"):
        if chunk.startswith('k'):
            p = int(chunk.replace("k", ""))
    return(k)

def get_n(filename):
    for chunk in filename.replace(".npy", "").split("_"):
        if chunk.startswith('n'):
            n = int(chunk.replace("n", ""))
    return(n)

def get_method(filename):
    for chunk in filename.replace(".npy", "").split("_"):
        if chunk.startswith('method'):
            method = chunk.replace("method", "")
    return(method)

dir_result = dir + "/result"
filelist=  [f for f in listdir(dir_result) if fnmatch.fnmatch(f, '*.npy')]

method_list = []
alpha_list = []
n_list = []
for fileitem in filelist:
    method_list.append( get_method(fileitem) )    
    alpha_list.append( get_alpha(fileitem))  
    n_list.append( get_n(fileitem))  
method_list = np.unique( np.array(method_list))
alpha_list = np.unique( np.array(alpha_list))
n_list = np.unique( np.array(n_list))

print(method_list)
print(alpha_list)
print(n_list)

file_dict  = dict(zip(method_list, [dict(
    zip(alpha_list, [dict(
        zip(n_list, ["" for p in n_list])
    ) for alpha in alpha_list])
) for method in method_list]))


for file_now  in filelist:
    method_now =  get_method(file_now)
    alpha_now =  get_alpha(file_now)
    n_now =  get_n(file_now)
    file_dict[method_now][alpha_now][n_now] = file_now


import numpy as np
import sqlite3
from datetime import datetime
con = sqlite3.connect('/Users/jmmoon/Documents/GitHub/LDPUts/experiment/LDP_minimax.db')
cursor_db = con.cursor()
```

```python
method_now = "genrrchi"
time_now = datetime.utcnow().strftime('%Y-%m-%d %H:%M:%S')


mechanism = "genrr"
statistic = "chi"

filelist_now = file_dict[method_now]
for priv_lev in filelist_now:
    for n in filelist_now[priv_lev]:
        data_now = np.load(dir_result + "/"+ filelist_now[priv_lev][n])

        for i in range(200):
            p_val_now = data_now[i,0]
            print(p_val_now)
            data_entry = (i+1, k, bump, priv_lev, int(n), statistic, mechanism, p_val_now, time_now)
            cursor_db.execute(
                "INSERT INTO ldp_disc_basic_comparison(rep, dim, bump, priv_lev, sample_size, statistic, mechanism, p_val, jobdate) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)", data_entry,
            )


con.commit()
```
