```python
import numpy as np
from os import listdir
import fnmatch

from os.path import isfile, join, isdir
import pandas as pd
```

```python
def get_k(filename):
    for chunk in filename.replace(".npy", "").split("_"):
        if chunk.startswith('k'):
            k = float(chunk.replace("k", ""))/10
    return(k)

def get_priv(filename):
    for chunk in filename.replace(".npy", "").split("_"):
        if chunk.startswith('priv'):
            priv = int(chunk.replace("priv", ""))
    return(priv)

def get_method(filename):
    for chunk in filename.replace(".npy", "").split("_"):
        if chunk.startswith('method'):
            method = chunk.replace("method", "")
    return(method)

def get_n(filename):
    for chunk in filename.replace(".npy", "").split("_"):
        if chunk.startswith('n'):
            n = int(chunk.replace("n", ""))
    return(n)
```

```python
filelist=  [f for f in listdir("./result/") if fnmatch.fnmatch(f, '*.npy')]

method_list = []
priv_list = []
n_list = []

for fileitem in filelist:
    method_list.append( get_method(fileitem) )    
    priv_list.append( get_priv(fileitem))  
    n_list.append( get_n(fileitem))  
method_list = np.unique( np.array(method_list))
priv_list = np.unique( np.array(priv_list))
n_list = np.unique( np.array(n_list))

print(method_list)
print(priv_list)
print(n_list)
```

```python



file_dict  = dict(zip(method_list, [dict(
    zip(priv_list, [dict(
        zip(n_list, ["" for n in n_list])
    ) for priv in priv_list])
) for method in method_list]))


for file_now  in filelist:
    method_now =  get_method(file_now)
    priv_now =  get_priv(file_now)
    n_now =  get_n(file_now)
    file_dict[method_now][priv_now][n_now] = file_now

experiment_name = "k_40"
df_list = dict(
    zip(method_list, [pd.DataFrame(columns=priv_list, index =  n_list) for method in method_list])
    )
```

```python

for method in method_list:
    for priv in priv_list:
        for n in n_list:
            try:
                data_now = np.load('result/'+file_dict[method][priv][n])
                power_now = (data_now<0.05).mean()
            except:
                power_now = 0
                print("fail")
            df_list[method].loc[n, priv] = power_now
    df_list[method].to_csv("./"+ experiment_name + "_method"+ method + ".csv" )
```
